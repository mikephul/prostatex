```python
import tensorflow as tf
import numpy as np
import math
# import timeit
import random
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.io as sio
from crop_data import *
```

```python
cropped_ADC = crop_data('data/ADC.mat', 'data/ADC_co.mat', height= 106, width=80, depth=19)
```

```python
def show_img(patient):
    nchannels = patient.shape[2]
    for c in range(nchannels):
        plt.subplot(1, 3, c + 1)
        plt.imshow(patient[:,:,c].astype('uint8'))
    plt.show()
```

## Show sample images

```python
coordinate_file = 'data/ADC_co.mat' 
matco = sio.loadmat(coordinate_file)
matco_key = coordinate_file.split('.')[0].split('/')[-1]
matco = matco[matco_key]

num_patients, height, width, depth = cropped_ADC.shape
samples = random.sample(range(1,num_patients), 4) 
num_samples = len(samples)
for i in range(num_samples):
    plt.subplot(1, num_samples, i + 1)
    slice_with_tumor = matco[samples[i]][0][2]
    plt.imshow(cropped_ADC[i,:,:,slice_with_tumor])
    plt.show()
```
